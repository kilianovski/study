```
from scipy.integrate import quad
import numpy as np
import pandas as pd
```

```
def normalProbabilityDensity(x):
  constant = 1.0 / np.sqrt(2*np.pi)
  return (constant * np.exp((-x**2) / 2.0) )
```

```
quad(normalProbabilityDensity, -10, 1)
```

```
def normal_CDF(x):
  res = quad(normalProbabilityDensity, np.NINF, x)
  return res
```

```
Z_table = pd.DataFrame(data=[], 
                       index=np.round(np.arange(0, 3.5, .1), decimals=2),
                       columns=np.round(np.arange(0, .1, .01), decimals=2))

for big_guy in Z_table.index:
  for small_guy in Z_table.columns:
    cdf_value, _ = normal_CDF(big_guy + small_guy)
    cdf_value = np.round(cdf_value, decimals=4)
    Z_table.loc[big_guy, small_guy] = cdf_value
    
Z_table
```
