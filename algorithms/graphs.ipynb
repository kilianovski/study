```python
ADJ_MOVES = [(0,+1), (+1,0), (0,-1), (-1,0)]

import collections

class Solution:
    def numIslands(self, grid) -> int:
        count = 0
        
        m,n = len(grid), len(grid[0])

        def explore_land(i,j):
            valid_moves = [(i+r, j+c) for r,c in ADJ_MOVES]
            valid_moves = [
                (i,j) for i,j in valid_moves if 
                i in range(m) 
                and j in range(n)
                and grid[i][j] == '1'
            ]

            return valid_moves


        def bfs(i,j):
            grid[i][j] = '0'
            q = collections.deque([(i,j)])
            c = 0
            while q:
                c += 1
                
                if c > 1_000: break
                
                i,j = q.popleft()
                grid[i][j] = '0'
                neighboring_land = explore_land(i,j)
                print(f'from ({i},{j}) {neighboring_land=}')
                q.extend(neighboring_land)

                # i,j = q.popleft()
                
                # neighboring_land = explore_land(i,j)
                
                # print(f'from ({i},{j}) {neighboring_land=}')

                # for (i,j) in neighboring_land:
                #     # grid[i][j] = '0'
                #     q.append((i,j))






        for i in range(m):
            for j in range(n):
                if grid[i][j] == '1':
                    count += 1
                    bfs(i,j)

        return count
```
