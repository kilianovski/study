## pivot!

```python
# Creating the DataFrame with the provided data
data = {
    "city": ["Jacksonville", "Jacksonville", "Jacksonville", "Jacksonville", "Jacksonville",
             "ElPaso", "ElPaso", "ElPaso", "ElPaso", "ElPaso"],
    "month": ["January", "February", "March", "April", "May", 
              "January", "February", "March", "April", "May"],
    "temperature": [13, 23, 38, 5, 34, 20, 6, 26, 2, 43]
}

weather = pd.DataFrame(data)
weather
```

```python
# Pivoting the DataFrame without the extra index layer
pivot_df = weather.pivot(index='month', columns='city', values='temperature')
# pivot_df.set_index('month')
pivot_df.reset_index(inplace=True)
pivot_df.columns.name = None  # Removing the extra layer index
pivot_df = pivot_df.set_index('month')
pivot_df
```
